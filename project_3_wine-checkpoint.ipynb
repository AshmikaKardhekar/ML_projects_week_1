```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#get data and insert column names along with it
wine_data = pd.read_csv('Downloads/wine.data',header=None,names=['wine-type','Alcohol','Malic acid','Ash','Alcalinity of ash',
                                                                 'Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols',
                                                                 'Proanthocyanins','Color intensity','Hue',
                                                                 'OD280/OD315 of diluted wines','Proline'])
wine_data.head()
```

```python
wine_data.shape
```

all the data seems to be continuous except for wine-type which is categorical

```python
sns.countplot(wine_data.iloc[:,0])
```

```python
#plotting the distribution of all features
for i in wine_data.iloc[:,1:].columns:
        plt.figure(i)
        sns.distplot(wine_data[i])
```

most of the data seems normally distributed

no null values

```python
#checking the correlation
plt.figure(figsize=(12,6))
sns.heatmap(wine_data.corr(),annot=True)
```

the data needs to be standardized to bring all the values on a common scale

```python
#scaling the data 
import sklearn
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scaled_wine_data=pd.DataFrame(scale.fit_transform(wine_data.iloc[:,1:]),columns=wine_data.iloc[:,1:].columns)
new_wine_data=pd.concat([scaled_wine_data,wine_data.iloc[:,0]],axis=1)
new_wine_data.head()
```

```python
#detecting the outliers
from  scipy.stats import zscore
z=abs(zscore(new_wine_data))
z[z>3]
```

```python
#removing outliers
wine = new_wine_data[(z<3).all(axis=1)]
```

```python
new_wine_data.shape,wine.shape
```

```python
#split target and input variables
x = wine.iloc[:,0:-1] #input variables
y = wine.iloc[:,-1] #target variable 
```

```python
x.shape,y.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import KFold,cross_val_score,GridSearchCV,train_test_split
```

```python
#choosing the best algorithm
#train and predict the data using cross-validation
model_list=[KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),
      AdaBoostClassifier(),GradientBoostingClassifier()]
for model in model_list:
    kfold=KFold(n_splits=10,random_state=None)
    results=cross_val_score(model,x,y,cv=kfold,scoring='accuracy')
    print(model)
    print(results.mean())
    print('\n')
```

RandomForestClassifier(rfc) and SupportVectorClassifier(svc) are almost equally working well.

COMPARING THE TUNED PERFORMANCE OF BOTH THE MODELS :

```python
#tuning the parameters of random forest
rfc=RandomForestClassifier()
param_grid={'n_estimators':range(100,500,100),
            'max_features':['auto','sqrt','log2'],
            'max_depth':[5,6,7,8,9,10],
            'criterion':['gini','entropy']}
rfc_cv=GridSearchCV(estimator=rfc,param_grid=param_grid,cv=5)
rfc_cv.fit(x,y)
```

```python
print('best params for RandomForestClassfier : ',rfc_cv.best_params_)
```

```python
print('best scores using tuned RandomForestClassifier : ',rfc_cv.best_score_)
```

```python
#tuning parameters of svc
svc=SVC()
params={'kernel':['rbf','poly'],
       'gamma':['scale','auto']}
svc_vc=GridSearchCV(estimator=svc,param_grid=params,cv=5)
svc_vc.fit(x,y)
```

```python
#best params for svc
print('best params for svc : ',svc_vc.best_params_)
```

```python
print('best score using tuned svc : ',svc_vc.best_score_)
```

best score using the tuned parameters of rfc is better than svc.


```python
#using best params of rfc for final training of all the data 
rfc_new=RandomForestClassifier(criterion='entropy',max_depth=7,max_features='sqrt',n_estimators=100)
rfc_new.fit(x,y)
```

```python
rfc_new.score(x,y)
```
